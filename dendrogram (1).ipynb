## **dendrogram**

A dendrogram (or clustering dendrogram) is a diagram that shows the hierarchical relationship between objects. It is most commonly created as an output from hierarchical clustering.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
data=np.array([[2],[14],[6],[7],[17]])
labels=['a(2)','b(14)','c(6)','d(7)','e(17)']
```

```python
plt.figure(figsize=(8,5))
plt.title("Student Hands on Example: Dendrogram")
dendrogram=sch.dendrogram(sch.linkage(data,method="single",metric="euclidean"),labels=labels)

plt.xlabel("Data Points")
plt.ylabel("Distance Between CLuster")
plt.grid(axis='y',linestyle="--",alpha=0.7)
plt.show()
```

```python
data=np.array([[1,1],[1.5,1.5],[5,5],[3,4],[4,4]])
labels=['A','B','C','D','E']
```

```python
plt.figure(figsize=(8,5))
plt.title("Dendrogram Single Linkage - closet pair")
dendrogram=sch.dendrogram(sch.linkage(data,method="single",metric="euclidean"),labels=labels)

plt.xlabel("Data Points")
plt.ylabel("Distance Between Cluster")
plt.grid(axis='y',linestyle=":",alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title("Dendrogram complete Linkage - closet pair")
dendrogram=sch.dendrogram(sch.linkage(data,method="complete",metric="euclidean"),labels=labels)

plt.xlabel("Data Points")
plt.ylabel("Distance Between Cluster")
plt.grid(axis='y',linestyle=":",alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title("Dendrogram ward Linkage - closet pair")
dendrogram=sch.dendrogram(sch.linkage(data,method="ward",metric="euclidean"),labels=labels)

plt.xlabel("Data Points")
plt.ylabel("Distance Between Cluster")
plt.grid(axis='y',linestyle=":",alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title("Dendrogram centroid Linkage - closet pair")
dendrogram=sch.dendrogram(sch.linkage(data,method="centroid",metric="euclidean"),labels=labels)

plt.xlabel("Data Points")
plt.ylabel("Distance Between Cluster")
plt.grid(axis='y',linestyle=":",alpha=0.7)
plt.show()
```

```python
from scipy.cluster.hierarchy import linkage, cophenet
from scipy.spatial.distance import pdist

methods = ["single", "complete", "average", "ward"]

for m in methods:
    Z = linkage(data, method=m)
    c, _ = cophenet(Z, pdist(data))
    print(f"{m}: {c:.4f}")
```
