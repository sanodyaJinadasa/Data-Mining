```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import BisectingKMeans
```

```python
file_path="/content/drive/MyDrive/Data Mining/ecommerce_customers.csv"
df=pd.read_csv(file_path)
```

```python
X=df.iloc[:,1:].values
```

```python
top_down_model=BisectingKMeans(n_clusters=3,random_state=42)
df['cluster']=top_down_model.fit_predict(X)
```

```python
plt.figure(figsize=(10, 6))

plt.scatter(X[:, 0], X[:, 1],
            c=df['cluster'], cmap='viridis', s=50)

plt.title("Top-Down Clustering (Bisecting KMeans)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

plt.colorbar(label="Cluster")
plt.show()
```

```python
plt.figure(figsize=(10, 6))

plt.scatter(df.iloc[:, 1], df.iloc[:, 2],
            c=df['cluster'], cmap='viridis', s=50)

plt.title("Customer Segments (Top-Down Approach)")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")

plt.show()
```

```python
centers = top_down_model.cluster_centers_

plt.scatter(X[:, 0], X[:, 1], c=df['cluster'], cmap='viridis', s=50)
plt.scatter(centers[:, 0], centers[:, 1],
            c='red', s=200, marker='X', label='Centroids')

plt.legend()
plt.show()
```
